<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
1. What Is a Drawdown?
</h3>

<p>
A <strong>drawdown</strong> measures the decline in the value of an investment from a historical peak to a subsequent trough.
</p>

<p>
Unlike volatility, which measures overall fluctuations, drawdown focuses specifically on <strong>downside risk</strong>.
</p>

<p>
It answers an important question for investors:
</p>

<p style="font-weight:bold;">
How much money could I lose before the investment recovers?
</p>

<p>
Drawdowns are widely used in portfolio management, hedge funds, and quantitative trading to evaluate risk and capital preservation.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
2. Mathematical Definition
</h3>

<p>
Let:
</p>

<ul>
<li>P<sub>t</sub> = portfolio value at time t</li>
<li>M<sub>t</sub> = running maximum value up to time t</li>
</ul>

<p style="font-weight:bold;">
M<sub>t</sub> = max(P<sub>1</sub>, P<sub>2</sub>, ..., P<sub>t</sub>)
</p>

<p>
The drawdown at time t is defined as:
</p>

<p style="font-weight:bold;">
DD<sub>t</sub> = (P<sub>t</sub> − M<sub>t</sub>) / M<sub>t</sub>
</p>

<p>
Drawdowns are typically expressed as negative percentages.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
4. Maximum Drawdown (MDD)
</h3>

<p>
The <strong>maximum drawdown</strong> is the largest observed loss from peak to trough during a given period.
</p>

<p style="font-weight:bold;">
MDD = min(DD<sub>t</sub>)
</p>

<p>
It represents the worst-case loss experienced by an investor during the investment horizon.
</p>

<p>
For example:
</p>

<ul>
<li>Maximum Drawdown = -25%</li>
</ul>

<p>
This means the portfolio lost 25% from its highest value before recovering.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
5. Drawdown Duration
</h3>

<p>
Drawdown duration measures the length of time it takes for an investment to recover from a drawdown.
</p>

<p>
It is defined as the time between:
</p>

<ul>
<li>the peak before the drawdown</li>
<li>the recovery to a new peak</li>
</ul>

<p>
Long drawdown durations can be psychologically difficult for investors even if the eventual recovery occurs.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
6. Why Drawdowns Matter
</h3>

<p>
Volatility measures fluctuations in both directions.
</p>

<p>
However, investors care more about <strong>losses than gains</strong>.
</p>

<p>
Drawdowns directly measure downside risk and capital loss.
</p>

<p>
This makes drawdown analysis especially important for:
</p>

<ul>
<li>hedge funds</li>
<li>portfolio managers</li>
<li>algorithmic trading strategies</li>
<li>risk management teams</li>
</ul>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
8. Drawdowns in Quantitative Trading
</h3>

<p>
Professional quantitative funds carefully monitor drawdowns when evaluating trading strategies.
</p>

<p>
Strategies with high average returns but extremely large drawdowns may be considered unacceptable due to risk exposure.
</p>

<p>
Risk limits are often imposed based on maximum allowable drawdowns.
</p>

<p>
For example:
</p>

<ul>
<li>Strategy shutdown if drawdown exceeds 20%</li>
<li>Position size reductions during drawdowns</li>
</ul>

</h3>

<p>
Drawdowns are often visualized as a time series showing the distance from the running peak.
</p>

<p>
This visualization helps analysts identify periods of stress and evaluate recovery times.
</p>

```python
# Loading the stocks of my portfolio
tickers = [
        "ADANIPORTS.NS",
        "RELINFRA.NS",
        "MON100.NS",
        "IOC.NS",
        "ONGC.NS",
        "IDEA.NS",
        "GOLDBEES.NS",
        "MAFANG.NS"
    ]
loader.load_multiple_tickers(tickers=tickers, interval="1d")
```

```python
# DrawDowns
data = loader.load_ticker_data("ADANIPORTS.NS", "1d")
rolling_max = data['Close'].cummax() # cumulative maximum(running peak)
drawdown = (data['Close']-rolling_max)/rolling_max # compute drawdown

max_drawdown = drawdown.min()

print(f"Maximum drawdown is {max_drawdown}")
plt.figure(figsize=(10,5))
drawdown.plot()
plt.title("Drawdown Curve for ADANIPORT.NS")
plt.ylabel("Drawdown")
plt.show()
```

1. Understanding Financial Risk
</h3>

<p>
In financial markets, <strong>risk</strong> refers to the uncertainty associated with future investment returns.
</p>

<p>
Investors are not only concerned about expected returns but also about the variability of those returns.
</p>

<p>
Two investments may generate the same average return, but the one with more stable returns is typically considered less risky.
</p>

<p>
Risk measurement is therefore a central problem in quantitative finance and portfolio management.
</p>

3. Measuring Risk with Volatility
</h3>

<p>
In quantitative finance, volatility is often used as a proxy for risk.
</p>

<p>
Higher volatility implies larger fluctuations in returns and therefore greater uncertainty about future outcomes.
</p>

<p>
Volatility is typically measured as the standard deviation of returns.
</p>

<p>
However, volatility alone does not capture the relationship between risk and return.
</p>

<p>
This leads to the concept of <strong>risk-adjusted performance</strong>.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
5. The Sharpe Ratio
</h3>

<p>
The <strong>Sharpe Ratio</strong> is one of the most widely used risk-adjusted performance metrics in finance.
</p>

<p>
It measures the excess return generated per unit of risk.
</p>

S = \frac{E[R_p] - R_f}{\sigma_p}

<p>
Where:
</p>

<ul>
<li>E[R<sub>p</sub>] = expected portfolio return</li>
<li>R<sub>f</sub> = risk-free rate</li>
<li>σ<sub>p</sub> = portfolio volatility</li>
</ul>

<p>
The Sharpe ratio therefore evaluates how efficiently an investment converts risk into return.
</p>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
7. Annualized Sharpe Ratio
</h3>

<p>
For daily returns, the Sharpe ratio is often annualized.
</p>

<p>
If returns are measured daily:
</p>

<p style="font-weight:bold;">
Sharpe<sub>annual</sub> = Sharpe<sub>daily</sub> × √252
</p>

<p>
This allows comparison between strategies operating at different frequencies.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
8. Limitations of the Sharpe Ratio
</h3>

<p>
Despite its popularity, the Sharpe ratio has several limitations.
</p>

<ul>
<li>assumes returns are normally distributed</li>
<li>penalizes upside volatility</li>
<li>may underestimate tail risk</li>
<li>can be distorted by extreme events</li>
</ul>

<p>
For this reason, other performance metrics are often used alongside it.
</p>

```python
returns = data.pct_change()

rf = 0.06 / 252 #risk free rate (6% annually for government bond)

excess_returns = returns - rf

sharpe_ratio = (excess_returns.mean() / returns.std()) * np.sqrt(252)


print(type(sharpe_ratio))
display(sharpe_ratio.head())
```

Interpretation :

| Sharpe Ratio | Meaning              |
| ------------ | -------------------- |
| < 0          | negative performance |
| 0 – 1        | poor                 |
| 1 – 2        | good                 |
| 2 – 3        | very good            |


The Sharpe ratio measures risk-adjusted performance by comparing excess return relative to volatility. A higher Sharpe ratio indicates better compensation for risk. By computing the Sharpe ratio for different assets, investors can compare performance on a risk-adjusted basis rather than simply comparing raw returns.
