<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
1. Prices vs Returns in Financial Modeling
</h3>

<p>
A fundamental concept in quantitative finance is the distinction between <strong>prices</strong> and <strong>returns</strong>.
</p>

<p>
Asset prices typically behave like a <strong>random walk with drift</strong>, meaning that future price movements depend on accumulated past movements and random shocks.
</p>

<p style="font-weight:bold;">
P<sub>t</sub> = P<sub>t−1</sub> + μ + ε<sub>t</sub>
</p>

<p>
where:
</p>

<ul>
<li>P<sub>t</sub> → price at time t</li>
<li>μ → drift (expected growth)</li>
<li>ε<sub>t</sub> → random noise</li>
</ul>

<p>
Because prices are non-stationary and trend over time, quantitative analysis is usually performed on <strong>returns</strong> instead.
</p>

<p>
Returns behave more like <strong>stationary stochastic processes</strong>.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
2. Definition of Financial Returns
</h3>

<p>
Returns measure the percentage change in price between two periods.
</p>

<p style="font-weight:bold;">
R<sub>t</sub> = (P<sub>t</sub> − P<sub>t−1</sub>) / P<sub>t−1</sub>
</p>

<p>
This is called the <strong>simple return</strong>.
</p>

<p>
Returns allow comparison between assets with different price levels and represent the actual gain or loss experienced by investors.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
3. Logarithmic Returns
</h3>

<p>
Logarithmic returns are widely used in quantitative finance due to their mathematical properties.
</p>

<p style="font-weight:bold;">
r<sub>t</sub> = ln(P<sub>t</sub> / P<sub>t−1</sub>)
</p>

<p>
Advantages of log returns:
</p>

<ul>
<li>additive through time</li>
<li>compatible with continuous-time finance models</li>
<li>simplifies stochastic calculus</li>
</ul>

<p>
For small returns:
</p>

<p style="font-weight:bold;">
log return ≈ simple return
</p>

5. Distribution of Returns
</h3>

<p>
Financial returns typically exhibit several stylized statistical properties:
</p>

<ul>
<li>heavy tails</li>
<li>negative skewness</li>
<li>volatility clustering</li>
<li>non-normal distributions</li>
</ul>

<p>
This means extreme price movements occur more frequently than predicted by a normal distribution.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
6. Volatility
</h3>

<p>
Volatility measures the degree of variability in asset returns.
</p>

<p style="font-weight:bold;">
σ = √Var(r)
</p>

<p>
In practice, volatility is estimated using the standard deviation of returns.
</p>

<p style="font-weight:bold;">
σ = √( (1 / (n−1)) Σ (r<sub>t</sub> − μ)² )
</p>

<p>
Volatility is the primary quantitative measure of <strong>financial risk</strong>.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
7. Annualized Volatility
</h3>

<p>
Volatility is often scaled to an annual measure using:
</p>

<p style="font-weight:bold;">
σ<sub>annual</sub> = σ<sub>daily</sub> × √252
</p>

<p>
where 252 represents the approximate number of trading days in a year.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
8. Volatility Clustering
</h3>

<p>
Financial time series exhibit a phenomenon known as <strong>volatility clustering</strong>.
</p>

<p>
Large price movements tend to be followed by large movements, while small movements tend to follow small movements.
</p>

<p>
This property violates the assumption of constant variance in classical statistical models.
</p>

<p>
Models such as <strong>ARCH</strong> and <strong>GARCH</strong> were developed to capture this behavior.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
9. Rolling Volatility
</h3>

<p>
Rolling volatility measures how volatility evolves over time.
</p>

<p style="font-weight:bold;">
σ<sub>t</sub> = std(r<sub>t−n</sub>, ..., r<sub>t</sub>)
</p>

<p>
Common windows include:
</p>

<ul>
<li>20 days (1 month)</li>
<li>60 days (quarter)</li>
<li>252 days (1 year)</li>
</ul>

<p>
Rolling volatility helps detect periods of market stress.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
1. What Is a Return?
</h3>

<p>
In finance, a <strong>return</strong> measures the relative change in the value of an asset over time.
Rather than examining absolute price differences, analysts measure percentage change so that assets with different price levels become comparable.
</p>

<p>
Instead of measuring:
</p>

<pre>
Price_t − Price_{t−1}
</pre>

<p>
we measure the <strong>relative change in price</strong>. This produces a normalized measure of gain or loss that can be compared across assets and time periods.
</p>

<p>
Several different definitions of returns are used in financial analysis depending on the context.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
2. Simple Return (Arithmetic Return)
</h3>

<p>
The most basic return definition is the <strong>simple return</strong>, also called the arithmetic return.
</p>

<p style="font-weight:bold;">
R<sub>t</sub> = (P<sub>t</sub> − P<sub>t−1</sub>) / P<sub>t−1</sub>
</p>

<p>
Where:
</p>

<ul>
<li>P<sub>t</sub> = price at time t</li>
<li>P<sub>t−1</sub> = price at previous time</li>
</ul>

<h4>Example</h4>

<pre>
Price: 100 → 105
Return = 5%
</pre>

<h4>Interpretation</h4>

<ul>
<li>Measures percentage gain or loss</li>
<li>Easy to interpret</li>
<li>Widely used in basic financial analysis</li>
</ul>

<h4>Common Use Cases</h4>

<ul>
<li>portfolio performance measurement</li>
<li>financial reporting</li>
<li>basic return calculations</li>
</ul>

<h4>Limitation</h4>

<p>
Simple returns cannot be easily aggregated across time.
</p>

<pre>
+10% followed by −10%
</pre>

<p>
does not produce zero total return due to compounding effects.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
3. Log Return (Continuously Compounded Return)
</h3>

<p>
The <strong>log return</strong> is the most widely used return measure in quantitative finance and financial modeling.
</p>

<p style="font-weight:bold;">
r<sub>t</sub> = ln(P<sub>t</sub> / P<sub>t−1</sub>)
</p>

<h4>Key Advantage</h4>

<p>
Log returns are <strong>additive across time</strong>.
</p>

<pre>
Total log return = log return day1 + log return day2 + ...
</pre>

<p>
This property greatly simplifies many statistical and stochastic models used in finance.
</p>

<h4>Common Applications</h4>

<ul>
<li>Black-Scholes option pricing</li>
<li>stochastic calculus models</li>
<li>volatility modeling</li>
<li>GARCH models</li>
<li>quantitative trading strategies</li>
</ul>

<h4>Important Insight</h4>

<p>
For small price movements:
</p>

<pre>
log return ≈ simple return
</pre>

<p>
This is why daily log returns and simple returns often appear nearly identical.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
4. Cumulative Return
</h3>

<p>
Cumulative return measures the <strong>total growth of an investment over a period of time</strong>.
</p>

<p style="font-weight:bold;">
R<sub>cum</sub> = (P<sub>T</sub> − P<sub>0</sub>) / P<sub>0</sub>
</p>

<h4>Example</h4>

<pre>
Initial price = 100
Final price = 130

Cumulative return = 30%
</pre>

<h4>Use Cases</h4>

<ul>
<li>portfolio performance tracking</li>
<li>strategy backtesting</li>
<li>long-term investment evaluation</li>
</ul>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
5. Annualized Return
</h3>

<p>
Returns observed over short time periods are often converted into yearly values to enable comparisons between investments.
</p>

<p style="font-weight:bold;">
Annualized Return ≈ Mean Daily Return × 252
</p>

<p>
The value 252 represents the approximate number of trading days in a year.
</p>

<h4>Applications</h4>

<ul>
<li>investment comparison</li>
<li>performance reporting</li>
<li>risk-adjusted metrics such as the Sharpe ratio</li>
</ul>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
6. Geometric Return
</h3>

<p>
The <strong>geometric return</strong> accounts for compounding effects and represents the average growth rate of an investment.
</p>

<p style="font-weight:bold;">
R<sub>geo</sub> = ( ∏ (1 + R<sub>t</sub>) )<sup>1/n</sup> − 1
</p>

<h4>Use Cases</h4>

<ul>
<li>long-term portfolio growth measurement</li>
<li>investment performance comparison</li>
<li>compound return analysis</li>
</ul>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
7. Excess Return
</h3>

<p>
Excess return measures how much an investment outperforms the <strong>risk-free rate</strong>.
</p>

<p style="font-weight:bold;">
R<sub>excess</sub> = R − R<sub>f</sub>
</p>

<p>
Where R<sub>f</sub> represents the return of a risk-free asset such as government treasury bills.
</p>

<h4>Use Cases</h4>

<ul>
<li>Sharpe ratio calculation</li>
<li>Capital Asset Pricing Model (CAPM)</li>
<li>risk-adjusted portfolio analysis</li>
</ul>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
8. Real Return
</h3>

<p>
Real return adjusts nominal investment returns for the effects of inflation.
</p>

<p style="font-weight:bold;">
R<sub>real</sub> = R<sub>nominal</sub> − Inflation
</p>

<h4>Use Cases</h4>

<ul>
<li>long-term wealth evaluation</li>
<li>macroeconomic investment analysis</li>
<li>retirement planning and purchasing power analysis</li>
</ul>

10. A Key Insight for Quantitative Modeling
</h3>

<p>
Financial prices often behave like a <strong>random walk</strong>.
However, returns tend to fluctuate around a relatively stable mean and are closer to stationary processes.
</p>

<p>
For this reason, most statistical models in quantitative finance operate on <strong>returns rather than raw prices</strong>.
</p>

<p>
Working with returns allows analysts to build models for volatility, correlation, risk estimation, and stochastic price dynamics.
</p>

```python
returns = data["Close"].pct_change().dropna()

returns.hist(bins=50)
```

```python
rolling_volatility = returns.rolling(30).std()

returns.rolling(30).std().plot()
```

* Interpretation from the rolling volatility

| Period    | Volatility Regime |
| --------- | ----------------- |
| May–June  | High volatility   |
| July–Jan  | Calm regime       |
| Feb–March | Volatility spike  |


The rolling 20-day volatility shows distinct volatility regimes. Early in the sample period volatility is elevated, followed by a prolonged calm period where volatility declines significantly. Toward the end of the sample, volatility rises sharply again, indicating increased market uncertainty. This behavior illustrates volatility clustering, a well-known stylized fact of financial markets.

```python
# calculate the log returns of the stock
log_returns = np.log(data["Close"] / data["Close"].shift(1)).dropna()
print(log_returns.head())
log_returns.describe()


# for daily data : log return ≈ simple return
log_returns.hist(bins=50)


```

```python
# calculate the cumulative returns of the stock
cumulative_returns = (1 + returns).cumprod() - 1 # this is percentage change wrt starting capital
print(cumulative_returns.tail())
cumulative_returns.plot()
```

```python
annual_return = returns.mean() * 252
print(f"Annualized Return: {annual_return}")
```

* Difference between annualized return and cumulative return: Annualized return is the average return per year, while cumulative return is the total return over a specific period. <h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
1. What Are Stylized Facts?
</h3>

<p>
In quantitative finance, <strong>stylized facts</strong> refer to empirical statistical properties of financial data that appear consistently across many markets, assets, and time periods.
</p>

<p>
These patterns are not derived from theoretical models but are instead discovered through observation of real market data.
</p>

<p>
Stylized facts are extremely important because they guide the development of financial models.
If a model fails to reproduce these empirical properties, it is generally considered unrealistic.
</p>

<p>
Many classical financial models assume idealized conditions, but real markets exhibit several statistical behaviors that must be accounted for.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
2. Heavy Tails (Fat Tails)
</h3>

<p>
Financial return distributions often display <strong>fat tails</strong>, meaning extreme price movements occur more frequently than predicted by a normal distribution.
</p>

<p>
Under a normal distribution, extreme events should be extremely rare.
However, financial markets experience large price shocks far more often than Gaussian models suggest.
</p>

<p>
Examples include:
</p>

<ul>
<li>market crashes</li>
<li>major macroeconomic announcements</li>
<li>financial crises</li>
<li>unexpected geopolitical events</li>
</ul>

<p>
Because of fat tails, risk models based solely on normal distributions often underestimate the probability of extreme losses.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
3. Volatility Clustering
</h3>

<p>
Another well-known property of financial markets is <strong>volatility clustering</strong>.
</p>

<p>
Periods of high volatility tend to be followed by high volatility, while calm periods tend to persist.
</p>

<p>
This means that market turbulence does not occur randomly; instead, volatility tends to cluster in time.
</p>

<p>
Examples include:
</p>

<ul>
<li>financial crises</li>
<li>economic recessions</li>
<li>major policy announcements</li>
</ul>

<p>
This phenomenon motivated the development of volatility models such as:
</p>

<ul>
<li>ARCH models</li>
<li>GARCH models</li>
</ul>

<p>
These models attempt to capture time-varying volatility observed in financial markets.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
4. Absence of Autocorrelation in Returns
</h3>

<p>
Daily financial returns generally exhibit <strong>very little autocorrelation</strong>.
</p>

<p>
This means that past returns provide very limited information about future returns.
</p>

<p>
In other words, short-term price movements are largely unpredictable and behave similarly to a random process.
</p>

<p>
This observation is closely related to the <strong>Efficient Market Hypothesis</strong>, which suggests that prices already reflect all available information.
</p>

<p>
However, although returns themselves show weak autocorrelation, <strong>volatility often displays strong autocorrelation</strong>.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
5. Leverage Effect
</h3>

<p>
The <strong>leverage effect</strong> refers to the tendency for negative returns to increase future volatility more than positive returns of the same magnitude.
</p>

<p>
When stock prices fall sharply, companies become more financially leveraged relative to their equity value, which increases perceived risk.
</p>

<p>
As a result, negative market shocks often lead to increased volatility in subsequent periods.
</p>

<p>
This asymmetric behavior is commonly observed in equity markets.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
6. Aggregational Gaussianity
</h3>

<p>
At very short time scales, financial return distributions often exhibit strong non-normality and heavy tails.
</p>

<p>
However, as the time horizon increases (for example from daily returns to monthly or yearly returns), the distribution tends to become more Gaussian.
</p>

<p>
This phenomenon is known as <strong>aggregational Gaussianity</strong>.
</p>

<p>
It arises due to the aggregation of many small independent price changes over longer time horizons.
</p>

2. The Normal Distribution Assumption
</h3>

<p>
The normal distribution is a symmetric bell-shaped probability distribution defined by two parameters:
</p>

<ul>
<li>mean (μ)</li>
<li>standard deviation (σ)</li>
</ul>

<p>
Its probability density function is given by:
</p>

<p style="font-weight:bold;">
f(x) = (1 / √(2πσ²)) exp(−(x − μ)² / 2σ²)
</p>

<p>
Many early financial models adopted this assumption because it simplifies mathematical analysis and allows closed-form solutions.
</p>

<p>
For example, the famous <strong>Black–Scholes option pricing model</strong> assumes log returns follow a normal distribution.
</p>

<li><strong>asymmetry</strong> — negative and positive returns may behave differently</li>
</ul>

<p>
These properties mean that extreme events such as financial crashes occur much more often than predicted by the normal distribution.

```python
# Normal distribution has a skewness of 0 and kurtosis of 3. If the returns have a skewness different from 0, it indicates that the distribution is not symmetric. If the kurtosis is different from 3, it indicates that the distribution has heavier or lighter tails than a normal distribution.

print(f"The skewness for the return is : {returns.skew()}")
print(f"The kurtosis for the return is : {returns.kurtosis()}")

# Interpretation of skewness and kurtosis:
# - If skewness is positive, it indicates that the distribution has a longer right tail, meaning there are more extreme positive returns than negative returns. If skewness is negative, it indicates that the distribution has a longer left tail, meaning there are more extreme negative returns than positive returns.
# - If kurtosis is greater than 3, it indicates that the distribution has heavier tails than a normal distribution, meaning there are more extreme returns (both positive and negative) than expected under a normal distribution. If kurtosis is less than 3, it indicates that the distribution has lighter tails than a normal distribution, meaning there are fewer extreme returns than expected under a normal distribution.
```

The return distribution exhibits positive skewness (~0.62), indicating a longer right tail and occasional large positive returns. The kurtosis (~5.12) is significantly higher than that of a normal distribution (3), indicating heavy tails.
```python
# Jarque-Bera test for normality
from scipy.stats import jarque_bera
jb_stat, jb_pvalue = stats.jarque_bera(returns)
print(f"Jarque-Bera Statistic: {jb_stat}, p-value: {jb_pvalue}")

# if the p-value is less than a significance level (e.g., 0.05), we reject the null hypothesis of normality, indicating that the returns do not follow a normal distribution.
if jb_pvalue < 0.05:
    print("Reject the null hypothesis of normality. The returns do not follow a normal distribution.")
else:
    print("Fail to reject the null hypothesis of normality. The returns may follow a normal distribution.")
    
```

</h3>

<p>
Volatility measures the <strong>degree of variation in asset returns over time</strong>.
It quantifies how much the price of an asset fluctuates around its average value.
</p>

<p>
In financial markets, volatility is often interpreted as a measure of <strong>risk or uncertainty</strong>.
Assets with high volatility experience large price movements, while assets with low volatility tend to move more gradually.
</p>

<p>
For example:
</p>

<ul>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
2. Mathematical Definition of Volatility
</h3>

<p>
Volatility is typically defined as the <strong>standard deviation of returns</strong>.
</p>

<p style="font-weight:bold;">
σ = √Var(R)
</p>

<p>
For a sample of returns:
</p>

<p style="font-weight:bold;">
σ = √( (1/(n−1)) Σ (Rᵢ − μ)² )
</p>

<p>
Where:
</p>

<ul>
<li>Rᵢ = individual return observations</li>
<li>μ = mean return</li>
<li>σ = volatility</li>
</ul>

<p>
Higher variance implies greater uncertainty in price movements.
</p>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
1. Implied Volatility Revisited
</h3>

<p>
Implied volatility is the level of volatility that makes an option pricing model reproduce the observed market price of an option.
</p>

<p>
Unlike historical volatility, which is calculated from past returns, implied volatility reflects the <strong>market's expectation of future volatility</strong>.
</p>

<p>
Because option traders actively buy and sell contracts based on expectations of future uncertainty, implied volatility can be interpreted as a forward-looking measure of risk.
</p>

<p>
Mathematically, implied volatility is the value of σ that solves the Black–Scholes pricing equation:
</p>

<p style="font-weight:bold;">
C = BS(S, K, r, T, σ)
</p>

<p>
Where:
</p>

<ul>
<li>S = underlying asset price</li>
<li>K = strike price</li>
<li>T = time to maturity</li>
<li>r = risk-free interest rate</li>
<li>σ = implied volatility</li>
</ul>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
2. Volatility Smile
</h3>

<p>
If the Black–Scholes model were perfectly accurate, implied volatility would be constant across all strike prices.
</p>

<p>
However, empirical observations show that implied volatility varies systematically with strike price.
</p>

<p>
When plotted against strike price, the implied volatility curve often forms a <strong>U-shaped pattern</strong> known as the <strong>volatility smile</strong>.
</p>

<p>
This phenomenon indicates that out-of-the-money options tend to have higher implied volatility than at-the-money options.
</p>

<p>
The volatility smile reflects market participants' concern about extreme price movements and tail risk.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
3. Volatility Skew
</h3>

<p>
In equity markets, the volatility pattern is often asymmetric rather than symmetric.
</p>

<p>
Out-of-the-money put options frequently exhibit higher implied volatility than call options.
</p>

<p>
This pattern is known as the <strong>volatility skew</strong> or <strong>volatility smirk</strong>.
</p>

<p>
The skew arises because investors are particularly concerned about large downward market moves.
</p>

<p>
As a result, demand for protective put options increases, pushing up their prices and implied volatility.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
4. Term Structure of Volatility
</h3>

<p>
Implied volatility also varies across option maturities.
</p>

<p>
The relationship between implied volatility and time to maturity is known as the <strong>term structure of volatility</strong>.
</p>

<p>
Typical shapes include:
</p>

<ul>
<li><strong>Upward sloping</strong> — long-term volatility expectations are higher.</li>
<li><strong>Downward sloping</strong> — short-term uncertainty dominates.</li>
<li><strong>Humped</strong> — volatility peaks at intermediate maturities.</li>
</ul>

<p>
Changes in the term structure often reflect macroeconomic uncertainty, earnings announcements, or major financial events.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
5. Volatility Surface
</h3>

<p>
Combining strike price and maturity produces a three-dimensional representation of implied volatility known as the <strong>volatility surface</strong>.
</p>

<p>
The volatility surface shows how implied volatility varies across:
</p>

<ul>
<li>different strike prices</li>
<li>different option maturities</li>
</ul>

<p>
This surface provides a comprehensive view of the market's expectations of future price dynamics.
</p>

<p>
Traders and quantitative analysts study the volatility surface to identify mispricing opportunities and to construct hedging strategies.
</p>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
1. Volatility Clustering
</h3>

<p>
One of the most important empirical observations in financial markets is the phenomenon of <strong>volatility clustering</strong>.
</p>

<p>
This property states that periods of high volatility tend to be followed by high volatility, while periods of low volatility tend to be followed by low volatility.
</p>

<p>
In other words, volatility exhibits <strong>persistence over time</strong>.
</p>

<p>
For example:
</p>

<ul>
<li>financial crises produce extended periods of high volatility</li>
<li>stable economic periods produce calm markets</li>
</ul>

<p>
This observation contradicts classical models that assume constant volatility.
</p>

<p>
As a result, financial economists developed models capable of capturing time-varying volatility.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
2. Conditional Heteroskedasticity
</h3>

<p>
The statistical concept underlying volatility clustering is known as <strong>conditional heteroskedasticity</strong>.
</p>

<p>
In standard regression models, the variance of errors is assumed to be constant:
</p>

<p style="font-weight:bold;">
Var(εₜ) = σ²
</p>

<p>
However, financial returns exhibit time-varying variance.
</p>

<p>
In such cases, the variance depends on past information:
</p>

<p style="font-weight:bold;">
Var(εₜ | past information)
</p>

<p>
This leads to models where volatility evolves dynamically through time.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
3. ARCH Model
</h3>

<p>
The <strong>Autoregressive Conditional Heteroskedasticity (ARCH)</strong> model was introduced by Robert Engle in 1982.
</p>

<p>
The basic idea is that today's volatility depends on the magnitude of past squared returns.
</p>

<p>
ARCH(q) model:
</p>

<p style="font-weight:bold;">
σ²ₜ = α₀ + α₁ε²ₜ₋₁ + α₂ε²ₜ₋₂ + ... + α_qε²ₜ₋q
</p>

<p>
Where:
</p>

<ul>
<li>σ²ₜ = conditional variance</li>
<li>εₜ = return shock</li>
<li>α parameters control how past shocks influence volatility</li>
</ul>

<p>
Large shocks increase future volatility.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
4. Limitations of ARCH
</h3>

<p>
Although ARCH models capture volatility clustering, they have several limitations.
</p>

<ul>
<li>require many lag terms</li>
<li>can become computationally inefficient</li>
<li>do not easily capture long-term volatility persistence</li>
</ul>

<p>
These limitations led to the development of the more flexible <strong>GARCH model</strong>.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
5. GARCH Model
</h3>

<p>
The <strong>Generalized ARCH (GARCH)</strong> model extends the ARCH framework by allowing past volatility to influence current volatility.
</p>

<p>
The most common specification is the GARCH(1,1) model.
</p>

<p style="font-weight:bold;">
σ²ₜ = ω + αε²ₜ₋₁ + βσ²ₜ₋₁
</p>

<p>
Where:
</p>

<ul>
<li>ω = long-run volatility level</li>
<li>α = impact of new shocks</li>
<li>β = persistence of past volatility</li>
</ul>

<p>
In financial markets, β is typically large, reflecting strong persistence in volatility.
</p>

---

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
6. Interpretation of GARCH Parameters
</h3>

<table style="width:100%; border-collapse: collapse; margin: 15px 0; font-size: 14px; border:1px solid #e2e8f0;">

<tr style="background-color:#2d3748; color:white;">
<th style="padding:10px;">Parameter</th>
<th style="padding:10px;">Interpretation</th>
</tr>

<tr>
<td style="padding:8px;">ω</td>
<td style="padding:8px;">baseline volatility level</td>
</tr>

<tr style="background:#f7fafc;">
<td style="padding:8px;">α</td>
<td style="padding:8px;">impact of recent shocks</td>
</tr>

<tr>
<td style="padding:8px;">β</td>
<td style="padding:8px;">volatility persistence</td>
</tr>

</table>

<p>
If α + β is close to 1, volatility shocks persist for long periods.
</p>
