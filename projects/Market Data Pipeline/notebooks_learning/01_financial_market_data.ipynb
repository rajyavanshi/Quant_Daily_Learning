<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
1. What is a Financial Asset?
</h3>

<p>
A financial asset is a tradable instrument that represents ownership, contractual claims on future cash flows, or exposure to market price movements. 
Financial assets are traded on organized exchanges or over-the-counter markets and form the fundamental building blocks of global capital markets.
</p>

<div style="display: grid; grid-template-columns: 1fr 1fr; gap: 15px; margin: 20px 0;">

<div style="background: #f8fafc; padding: 15px; border-radius: 6px; border: 1px solid #edf2f7;">
<strong>Stocks & Equity</strong><br>
<span style="font-size: 14px;">
Ownership shares in a corporation. Investors may earn capital gains and dividends.
Example: Apple (AAPL), Tesla (TSLA).
</span>
</div>

<div style="background: #f8fafc; padding: 15px; border-radius: 6px; border: 1px solid #edf2f7;">
<strong>Indices & ETFs</strong><br>
<span style="font-size: 14px;">
Indices measure the performance of a basket of assets (S&P 500, NASDAQ).
ETFs allow investors to trade those indices like individual securities.
</span>
</div>

</div>

<p>
Other important categories include commodities, currencies, derivatives, and cryptocurrencies.
In quantitative finance, each asset is modeled as a <strong>stochastic price process evolving through time</strong>.
</p>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
4. Adjusted Close Price
</h3>

<p>
Raw closing prices do not always reflect the true economic value of an asset because of corporate actions.
</p>

<ul>
<li><strong>Stock Splits</strong> – change the price but not investor wealth</li>
<li><strong>Dividends</strong> – reduce the stock price when paid</li>
</ul>

<p>
To correct these distortions, financial datasets include the column <strong>Adjusted Close</strong>.
</p>

<p>
Quantitative calculations such as returns, volatility, and portfolio optimization should always use:
</p>

<p style="font-weight:bold;">
Adjusted Close
</p>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
5. Mathematical Representation of Price
</h3>

<p>
In quantitative finance, asset prices are modeled as stochastic processes.
Let:
</p>

<p>
P<sub>t</sub> = price of an asset at time t
</p>

<p>
The sequence
</p>

<p>
P₁, P₂, P₃, ... , Pₜ
</p>

<p>
forms a discrete-time stochastic process describing the evolution of the market.
</p>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
6. Log Returns
</h3>

<p>
Instead of analyzing raw prices, quantitative models typically analyze <strong>returns</strong>.
</p>

<p>
The logarithmic return is defined as:
</p>

<p style="font-weight:bold;">
r<sub>t</sub> = log(P<sub>t</sub> / P<sub>t-1</sub>)
</p>

<p>
Log returns are preferred because they:
</p>

<ul>
<li>are additive through time</li>
<li>simplify mathematical modeling</li>
<li>are compatible with continuous-time finance models</li>
</ul>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
7. Time Indexing
</h3>

<p>
Financial time series are indexed by trading time rather than calendar time.
</p>

<p>
This means that weekends and holidays appear as gaps in the dataset.
These gaps are normal and do not indicate missing data.
</p>

```python
import pandas as pd
import yfinance as yf

def get_stock_data(ticker, start_date, end_date):
    """
    Fetch historical stock data for a given ticker symbol and date range.

    Parameters:
    ticker (str): The stock ticker symbol (e.g., 'AAPL' for Apple Inc.).
    start_date (str): The start date for the data in 'YYYY-MM-DD' format.
    end_date (str): The end date for the data in 'YYYY-MM-DD' format.

    Returns:
    pd.DataFrame: A DataFrame containing the historical stock data.
    """
    try:
        # Fetch the stock data using yfinance
        stock_data = yf.download(ticker, start=start_date, end=end_date)
        return stock_data
    except Exception as e:
        print(f"An error occurred while fetching stock data: {e}")
        return pd.DataFrame()
if __name__ == "__main__":
    # Example usage
    ticker = "AAPL"
    
    start_date = "2020-01-01"
    end_date = "2021-01-01"
    
    data = get_stock_data(ticker, start_date, end_date)
    display(data.head())
```

1. Purpose of `.describe()`
</h3>

<p>
The <strong>.describe()</strong> function in Pandas provides a concise statistical summary of numerical variables within a dataset.
It is one of the first exploratory tools used in data analysis to understand the general structure of the data before building statistical or machine learning models.
</p>

<p>
For financial datasets, `.describe()` helps analysts quickly inspect important characteristics such as:
</p>

<ul>
<li>average return of an asset</li>
<li>volatility of the time series</li>
<li>presence of extreme observations</li>
<li>distribution symmetry or skewness</li>
<li>data completeness</li>
</ul>

<p>
Because financial markets generate massive amounts of time-series data, summary statistics act as a <strong>first diagnostic snapshot of market behavior</strong>.
</p>

4. Mathematical Interpretation
</h3>

<p>
Let a dataset contain observations:
</p>

<p>
x₁, x₂, x₃, … , xₙ
</p>

<p>
The mean is defined as:
</p>

<p style="font-weight:bold;">
μ = (1/n) Σ xᵢ
</p>

<p>
The standard deviation measures dispersion around the mean:
</p>

<p style="font-weight:bold;">
σ = √((1/(n−1)) Σ (xᵢ − μ)²)
</p>

<p>
Quartiles divide the ordered dataset into four equal segments.
</p>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
5. Interquartile Range and Outliers
</h3>

<p>
The interquartile range is defined as:
</p>

<p style="font-weight:bold;">
IQR = Q3 − Q1
</p>

<p>
Outliers are often defined as values outside:
</p>

<p style="font-weight:bold;">
[ Q1 − 1.5 × IQR , Q3 + 1.5 × IQR ]
</p>

<p>
In finance, extreme outliers often correspond to:
</p>

<ul>
<li>earnings announcements</li>
<li>macroeconomic shocks</li>
<li>market crashes</li>
<li>short squeezes</li>
</ul>

6. Interpreting `.describe()` in Practice
</h3>

<p>
Professional analysts can extract multiple insights instantly from the summary table.
</p>

<ul>
<li><strong>Missing Data:</strong> Compare dataset size with the count value.</li>
<li><strong>Skewness:</strong> Compare mean and median.</li>
<li><strong>Outliers:</strong> Observe the gap between max and the upper quartile.</li>
<li><strong>Volatility:</strong> Assess the magnitude of the standard deviation.</li>
<li><strong>Scale:</strong> Check whether values are realistic.</li>
</ul>

<p>
For example, if:
</p>

<pre>
mean = 0.001
std = 0.03
</pre>

<p>
The volatility is <strong>30 times larger than the mean</strong>, indicating a highly volatile asset.
</p>

```python
import pandas as pd
import numpy as np

def data_diagnostic(df):

    report = pd.DataFrame()

    report["dtype"] = df.dtypes
    report["count"] = df.count()
    report["missing"] = df.isna().sum()
    report["missing_%"] = (df.isna().sum() / len(df)) * 100

    report["mean"] = df.mean(numeric_only=True)
    report["std"] = df.std(numeric_only=True)
    report["min"] = df.min(numeric_only=True)
    report["25%"] = df.quantile(0.25, numeric_only=True)
    report["median"] = df.median(numeric_only=True)
    report["75%"] = df.quantile(0.75, numeric_only=True)
    report["max"] = df.max(numeric_only=True)

    report["skew"] = df.skew(numeric_only=True)
    report["kurtosis"] = df.kurtosis(numeric_only=True)

    return report

data_diagnostic(data)
```

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
1. Correlation Analysis
</h3>

<p>
Correlation analysis examines the statistical relationship between different variables in a dataset. 
For financial market data, this typically includes variables such as:
</p>

<ul>
<li>Open price</li>
<li>High price</li>
<li>Low price</li>
<li>Close price</li>
<li>Trading volume</li>
</ul>

<p>
The correlation coefficient ranges between <strong>-1 and 1</strong>.
</p>

<ul>
<li><strong>+1</strong> → perfect positive correlation</li>
<li><strong>0</strong> → no relationship</li>
<li><strong>-1</strong> → perfect negative correlation</li>
</ul>

<p>
In financial datasets, prices such as Open, High, Low, and Close usually exhibit very strong positive correlations because they originate from the same trading process.
</p>

<p>
Volume, however, often behaves differently and may show weaker correlations with prices.
</p>


<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
2. Volatility Analysis
</h3>

<p>
Volatility represents the degree of variation in an asset’s price over time.
It is commonly measured using the <strong>standard deviation of returns</strong>.
</p>

<p style="font-weight:bold;">
Volatility = Standard Deviation of Returns
</p>

<p>
In finance, volatility is a proxy for <strong>risk</strong>.
</p>

<ul>
<li>High volatility → large price fluctuations</li>
<li>Low volatility → stable price movement</li>
</ul>

<p>
Many financial models such as the Black–Scholes option pricing model rely heavily on volatility estimation.
</p>

3. Moving Averages
</h3>

<p>

</p>

<ul>

<li><strong>200-day moving average</strong></li>

</p>

<p>
When the short-term average crosses below the long-term average, it is known as a <strong>Death Cross</strong>, often interpreted as a bearish signal.
</p>

4. Return Analysis
</h3>

<p>
Returns measure the percentage change in price between two periods.
</p>

<p style="font-weight:bold;">
Return = (P<sub>t</sub> − P<sub>t−1</sub>) / P<sub>t−1</sub>
</p>

<p>
Analyzing return distributions helps researchers understand:
</p>

<ul>
<li>average profitability</li>
<li>risk exposure</li>
<li>extreme market events</li>
</ul>

<p>
In quantitative finance, returns are preferred over raw prices because they are scale-independent.
</p>

<h3 style="color: #2c5282; font-family: 'Crimson Text', serif; font-size: 22px; border-left: 4px solid #2c5282; padding-left: 15px;">
11. Skewness and Kurtosis Analysis
</h3>

<p>
Higher-order moments such as skewness and kurtosis describe the shape of return distributions.
</p>

<p>
<strong>Skewness</strong> measures asymmetry.
</p>

<ul>
<li>Positive skew → extreme positive returns more likely</li>
<li>Negative skew → extreme negative returns more likely</li>
</ul>

<p>
<strong>Kurtosis</strong> measures tail heaviness.
</p>

<ul>
<li>Kurtosis &gt; 3 → heavy tails</li>
<li>Kurtosis = 3 → normal distribution</li>
<li>Kurtosis &lt; 3 → light tails</li>
</ul>

<p>
Financial return distributions typically exhibit <strong>high kurtosis</strong>, meaning extreme events occur more frequently than predicted by normal distributions.
</p>

```python
# kurtois and skewness

def conclusion(df, column_name):
    skewness = df[column_name].skew()
    kurtosis = df[column_name].kurtosis()

    if skewness > 0:
        skewness_conclusion = "The distribution is positively skewed (right-skewed)."
    elif skewness < 0:
        skewness_conclusion = "The distribution is negatively skewed (left-skewed)."
    else:
        skewness_conclusion = "The distribution is symmetric."

    if kurtosis > 3:
        kurtosis_conclusion = "The distribution has heavy tails (leptokurtic)."
    elif kurtosis < 3:
        kurtosis_conclusion = "The distribution has light tails (platykurtic)."
    else:
        kurtosis_conclusion = "The distribution has normal tails (mesokurtic)."

    report = {
        "skewness": skewness,
        "skewness_conclusion": skewness_conclusion,
        "kurtosis": kurtosis,
        "kurtosis_conclusion": kurtosis_conclusion
    }
    
    return report

print(data.columns) # to check the column names and their format
# data is in multiindex format, so we need to use the correct column name to access the "Close" column
# the conclusion function expects a DataFrame and a column name, so we need to pass the DataFrame and the column name as arguments

conclusion_report = conclusion(data, ( 'Close', 'ADANIPORTS.NS'))
print("Skewness:", conclusion_report["skewness"])
print("Skewness Conclusion:", conclusion_report["skewness_conclusion"])
print("Kurtosis:", conclusion_report["kurtosis"])
print("Kurtosis Conclusion:", conclusion_report["kurtosis_conclusion"])

```
