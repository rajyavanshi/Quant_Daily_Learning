## Return Distribution Diagnostics for a Volatile Asset (‘RDD-VA’ Study)

### Business / Company Context

A mid-sized quant research firm is building a lightweight diagnostic toolkit to quickly assess whether an asset’s daily returns behave according to classical models (Gaussian) or show heavy-tailed, unstable behaviour.

This toolkit can be further used by:

* risk teams (VaR estimation)

* trading teams (volatility prediction)

* quant research (model validation)

We are assigned a mini-project to create the first prototype using real market data.

### Project Aim

To analyze real-world return behaviour for any chosen financial asset and determine:

whether its returns follow a normal distribution or not

which alternative distribution fits better

what the tail-risk characteristics look like

### Problem Statement

Most trading models still assume Gaussian returns, which leads to severe underestimation of extreme movements (crashes/spikes).
Our task is to collect real price data, compute returns, and perform quantitative diagnostics to check:

<i>“Does the asset exhibit heavy tails, skewness, kurtosis, and rare-event clustering?”</i>
